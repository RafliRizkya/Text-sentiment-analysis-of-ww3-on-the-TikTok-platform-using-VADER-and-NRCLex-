# Data Cleaning

```python
#menghapus karakter yang bukan huruf, seperti emoji, simbol yang aneh
def remove_non_ascii(text):
    return text.encode('ascii', 'ignore').decode('ascii')

text['text'] = text['text'].apply(remove_non_ascii)
```

```python
import re

def clean_text(text):

    """
    Membersihkan teks dari karakter yang tidak relevan untuk analisis.

    Proses pembersihan mencakup:
    - Mengubah seluruh teks menjadi huruf kecil (lowercase)
    - Menghapus URL
    - Menghapus kata yang diawali dengan tanda '#' (hashtag)
    - Menghapus simbol selain huruf, angka, dan spasi
    - Menghapus spasi ganda

    Parameter:
    - text (str): Teks mentah yang akan dibersihkan

    Return:
    - str: Teks yang telah dibersihkan
    """
    
    text = text.lower()                                 # lowercase
    text = re.sub(r'http\S+', '', text)                 # hapus URL
    text = re.sub(r'#\w+', '', text)                    # hapus kata yang diawali dengan #
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)          # hapus simbol lain (kecuali huruf, angka, spasi)
    text = re.sub(r'\s+', ' ', text).strip()            # hapus spasi ganda
    return text

text['text'] = text['text'].apply(clean_text)
text = text[text['text'].str.strip() != '']
```

```python
# Mengunduh dan memuat daftar stopwords bahasa Inggris untuk menghapus kata umum yang kurang bermakna
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
```

```python
from collections import Counter

# Tokenisasi kata
words = text['text'].str.split().explode()

# Hapus stopwords
filtered_words = words[
    ~words.str.lower().isin(stop_words)]

# Hitung frekuensi kata
word_freq = Counter(filtered_words.str.lower())  # lowercase semua
```

```python
from wordcloud import WordCloud

def create_wc(data,capt):

    """
    Membuat visualisasi WordCloud berdasarkan frekuensi kata.

    Parameters:
    - data (dict): Kamus yang berisi pasangan kata dan jumlah kemunculannya.
    - capt (str): Nama file (dengan format .png atau .jpg) untuk menyimpan gambar WordCloud.

    Output:
    - Menampilkan WordCloud menggunakan matplotlib.
    - Menyimpan hasil WordCloud ke file dengan nama sesuai parameter 'capt'.
    """

    text_wc = WordCloud(width=1600, height=800, background_color='white', max_font_size= 200).generate_from_frequencies(data)

    plt.figure(figsize=(12, 6))
    plt.imshow(text_wc, interpolation='bilinear')
    plt.axis('off')

    plt.savefig(capt, dpi = 600, bbox_inches = 'tight')
    plt.tight_layout()
    plt.show()
```

```python
#Visualisasi Kata yang Paling Sering Muncul
create_wc(word_freq, 'Visualisasi Kata yang Paling Sering Muncul')
```

# Analisis Sentiment Positif, Netral, dan Negatif Menggunakan VADER

```python
# Mengklasifikasikan sentimen teks menjadi Positif, Negatif, atau Netral menggunakan VADER
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()

def get_sentiment(text):
    score = analyzer.polarity_scores(text)
    compound = score['compound']
    if compound >= 0.05:
        return 'Positive'
    elif compound <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

text['sentiment'] = text['text'].apply(get_sentiment)
```

```python
# Mengekstrak skor sentimen VADER (positif, netral, negatif, dan compound) ke dalam kolom terpisah
text['scores'] = text['text'].apply(analyzer.polarity_scores)
text['pos'] = text['scores'].apply(lambda x: x['pos'])
text['neu'] = text['scores'].apply(lambda x: x['neu'])
text['neg'] = text['scores'].apply(lambda x: x['neg'])
text['compound'] = text['scores'].apply(lambda x: x['compound'])

text = text.drop(columns= 'scores')
```

```python
def create_dict(data, col_val):

    """
    Fungsi ini menghasilkan dictionary frekuensi kata dari data berdasarkan nilai sentimen tertentu.
    
    Parameters:
        data (DataFrame): DataFrame yang berisi kolom 'text' dan 'sentiment'.
        col_val (str): Nilai sentimen yang ingin difilter, seperti 'Positive', 'Negative', atau 'Neutral'.
    
    Returns:
        Counter: Objek Counter yang berisi frekuensi kemunculan setiap kata pada sentimen yang dipilih.
    """
        
    sentiment = data[data['sentiment'] == col_val]
    sentiment = sentiment[sentiment['text'].notna()]
    sentiment_freq = Counter(sentiment['text'].str.split().explode().dropna())
    return sentiment_freq

negative_freq = create_dict(text, 'Negative')
netral_freq = create_dict(text, 'Neutral')
positive_freq = create_dict(text, 'Positive')
```

```python
#buat ulang stopwords
stopwords = set([
    'would', 'the', 'in', 'had', 'between', 'on', 'read', 'tulis', 'komentar', 'comment', 'this', 'were', 'on'\
    'what', 'if', 'is', 'that', 'so', 'and', 'after', 'was', 'saying', 'only', 'do', 'be', 'its', 'of', 'it'\
    'has', 'can', 'your', 'an', 'bio', 'a', 'to', 'will', 'you', 'yea', 'yup'
])

# Terapkan filtering stopwords ke Counter
def remove_stopwords(counter, stopwords):
    return Counter({word: count for word, count in counter.items() if word.lower() not in stopwords})

# Terapkan ke masing-masing freq dict
negative_freq = remove_stopwords(negative_freq, stopwords)
netral_freq   = remove_stopwords(netral_freq, stopwords)
positive_freq = remove_stopwords(positive_freq, stopwords)
```

# Emotion Extraction Menggunakan NRCLex

```python
from nrclex import NRCLex

def extract_emotion(text):

    """
    Fungsi ini mengambil emosi dominan dari sebuah teks menggunakan library NRCLex.
    Jika emosi ditemukan, maka akan mengembalikan emosi teratas (paling dominan).
    Jika tidak ditemukan emosi, akan mengembalikan 'none'.
    Jika terjadi error dalam proses, akan mengembalikan 'error'.
    
    Parameter:
        text (str): Kalimat atau teks yang ingin dianalisis.

    Return:
        str: Nama emosi dominan atau 'none' / 'error'.
    """

    try:
        emotion = NRCLex(text)
        top = emotion.top_emotions
        if top:
            return top[0][0]
        else:
            return 'none'
    except:
        return 'error'
    
```

```python
text['dominant_emotion'] = text['text'].apply(lambda x: extract_emotion(x[:300]))
```
